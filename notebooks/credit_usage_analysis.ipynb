```python
import pandas as pd

# Load the cleaned dataset
df = pd.read_csv(r"C:\Users\Admin\Desktop\-Credit-Card-Customer-Analysis\data\credit_card_data_csv.csv")

# Print the column names to verify
print("Column names in the dataset:")
print(df.columns)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt



# Analyze credit usage
plt.figure(figsize=(18, 6))

# Credit Limit vs Customer Age
plt.subplot(1, 3, 1)
sns.scatterplot(x='Customer_Age', y='Credit_Limit', data=df)
plt.title('Credit Limit vs Age')
plt.xlabel('Age')
plt.ylabel('Credit Limit')

# Balance vs Income Category
plt.subplot(1, 3, 2)
sns.scatterplot(x='Income_Category', y='Total_Revolving_Bal', data=df)
plt.title('Balance vs Income Category')
plt.xlabel('Income Category')
plt.ylabel('Balance')
plt.xticks(rotation=45, ha='right')

# Credit Limit by Marital Status
plt.subplot(1, 3, 3)
sns.boxplot(x='Marital_Status', y='Credit_Limit', data=df)
plt.title('Credit Limit by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Credit Limit')
plt.xticks(rotation=45, ha='right')

plt.tight_layout(pad=3.0)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the clustered dataset
df = pd.read_csv(r"C:\Users\Admin\Desktop\-Credit-Card-Customer-Analysis\data\clustered_credit_card_data.csv")

# Analyze each cluster
for cluster in df['cluster'].unique():
    cluster_data = df[df['cluster'] == cluster]
    print(f'Cluster {cluster} Statistics:')
    print(cluster_data.describe())
    print('\n')
```

```python
# Ensure correct column names
print("Column names in the dataset:")
print(df.columns)

# Visualize clusters
plt.figure(figsize=(18, 6))
```

```python
# Age by Cluster
plt.subplot(1, 3, 1)
sns.boxplot(x='cluster', y='Customer_Age', data=df)
plt.title('Age by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Age')
```

```python

# Credit Limit by Cluster
plt.subplot(1, 3, 2)
sns.boxplot(x='cluster', y='Credit_Limit', data=df)
plt.title('Credit Limit by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Credit Limit')
```

```python

# Balance by Cluster
plt.subplot(1, 3, 3)
sns.boxplot(x='cluster', y='Total_Revolving_Bal', data=df)
plt.title('Balance by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Balance')

plt.tight_layout()
plt.show()
```

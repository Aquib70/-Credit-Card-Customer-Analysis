```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
print("Loading the cleaned dataset...")
df = pd.read_csv(r"C:\Users\Admin\Desktop\-Credit-Card-Customer-Analysis\data\cleaned_credit_card_data.csv")
print("Dataset loaded successfully.")
```

```python
# Analyze demographics
print("Extracting demographic information...")
demographics = df[['Customer_Age', 'Income_Category', 'Marital_Status']]
print("Demographics extracted successfully.")
```

```python
# Visualize distributions
print("Visualizing distributions...")
plt.figure(figsize=(15, 5))

```

```python
# Age distribution
print("Plotting age distribution...")
plt.subplot(1, 3, 1)
sns.histplot(demographics['Customer_Age'], kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Income Category distribution
print("Plotting income category distribution...")
plt.subplot(1, 3, 2)
sns.countplot(x='Income_Category', data=demographics, order=demographics['Income_Category'].value_counts().index)
plt.title('Income Category Distribution')
plt.xlabel('Income Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Marital Status distribution
print("Plotting marital status distribution...")
plt.subplot(1, 3, 3)
sns.countplot(x='Marital_Status', data=demographics)
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.tight_layout(pad=3.0)
print("Displaying plots...")
plt.show()
```
